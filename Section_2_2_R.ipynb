```r
v<-c(20,40,60,80,100)
R<-c(8.3,16.7,25.0,33.3,41.7)
B<-c(2.3,9.0,20.2,35.8,55.9)
```

```r
# R vs v
```

```r
plot(v,R,col="blue", main="Reaction Time Distance vs. Speed",xlab="Speed (km/hr)", ylab="Reaction Time Distance (Meters)")
```

```r
relation<-lm(R~I(v))
print(relation)
summary(relation)
```

```r
v_seq <- seq(from=20,to=100, by=0.1)
new_relation<-list(0)
for (i in 1:length(v_seq)) {
    v_new <- data.frame(v=v_seq[i])
    new_relation[i]<-predict(relation, v_new)
}
plot(v,R,col="blue", main="Reaction Time Distance vs. Speed",xlab="Speed (km/hr)", ylab="Reaction Time Distance (Meters)",lines(v_seq, new_relation))
```

```r
RSE_Percent = 0.03651/mean(R) * 100
print (RSE_Percent)
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```

```r
# B vs v
```

```r
plot(v,B,col="blue", main="Braking Distance vs. Speed",xlab="Speed (km/hr)", ylab="Braking Distance (Meters)")
```

```r
relation2<-lm(B~I(v)+I(v^2))
print(relation2)
summary(relation2)
```

```r
v_seq2 <- seq(from=20,to=100, by=0.1)
new_relation2<-list(0)
for (i in 1:length(v_seq2)) {
    v_new2 <- data.frame(v=v_seq2[i])
    new_relation2[i]<-predict(relation2, v_new2)
}
plot(v,B,col="blue", main="Braking Distance vs. Speed",xlab="Speed (km/hr)", ylab="Braking Distance (Meters)",lines(v_seq2, new_relation2))
```

```r
RSE_Percent = 0.0169/mean(B) * 100
print (RSE_Percent)
```

```r
confint(relation2, level=0.95)
```

```r
plot(fitted(relation2), residuals(relation2))
```
