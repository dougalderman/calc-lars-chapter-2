```r
t<-c(4,5,6,7,8,9)
h<-c(1773,1890,2017,2135,2234,2330)
p<-c(293,296,299,302,305,307)
```

```r
# h vs t
```

```r
plot(t,h,col="blue",main="Health Care Expenditures vs. Year",xlab="Year (t = 4 correspoding to 2004)",ylab="Health Care Expenditures (Billions of $)")
```

```r
relation<-lm(h~I(t))
print(relation)
summary(relation)
```

```r
t_seq <- seq(from=4,to=9, by=0.1)
new_relation<-list(0)
for (i in 1:length(t_seq)) {
    t_new <- data.frame(t=t_seq[i])
    new_relation[i]<-predict(relation, t_new)
}
plot(t,h,col="blue",main="Health Care Expenditures vs. Year",xlab="Year (t = 4 correspoding to 2004)",ylab="Health Care Expenditures (Billions of $)",lines(t_seq, new_relation))
```

```r
RSE_Percent = 12.8/mean(h) * 100
print (RSE_Percent)
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```

```r
# p vs t
```

```r
plot(t,p,col="blue",main="U.S. Population vs. Year",xlab="Year (t = 4 correspoding to 2004)",ylab="U.S. Population (in millions)")
```

```r
relation2<-lm(p~I(t))
print(relation2)
summary(relation2)
```

```r
t_seq2 <- seq(from=4,to=9, by=0.1)
new_relation2<-list(0)
for (i in 1:length(t_seq2)) {
    t_new2 <- data.frame(t=t_seq2[i])
    new_relation2[i]<-predict(relation2, t_new2)
}
plot(t,p,col="blue",main="U.S. Population vs. Year",xlab="Year (t = 4 correspoding to 2004)",ylab="U.S. Population (millions)",lines(t_seq2, new_relation2))
```

```r
RSE_Percent = 0.345/mean(p) * 100
print (RSE_Percent)
```

```r
confint(relation2, level=0.95)
```

```r
plot(fitted(relation2), residuals(relation2))
```
